# Population distribution, Sample distribution and Sampling distributions 

###  This excercise will give clarity on the difference between Population, Sample and Sampling distributions

```r
# Load required libraries 

suppressWarnings(library(testthat))
suppressWarnings(library(digest))
suppressWarnings(library(infer))
suppressWarnings(library(tidyverse))
suppressWarnings(library(repr))
suppressWarnings(library(gridExtra))
```

Let us create a synthetic dataset to investigate how much time on an average the people of a locality spend excercising.
The below code creates a virually simulated population data (`excercise_data`) where the response variable is the number of hours spent per week on excercising by each of the 1,000,000 residents. 
We also create an `resident_id` column with a unique number for each person in the data set.
The hours of excercising come from two separate normal distributions with different mean and standard deviations. 

```r
# create virtual population of size 1000000 from a locality 

set.seed(1234)

n = 1000000;

y1 = rnorm(n, 4, 2) ; 
y2 = rnorm(n, 10, 1)

w = rbinom(n, 1, .4)
X = w*y1 + (1-w)*y2

excercise_data <- tibble(resident_id = seq(from = 1, to = n, by = 1),
                                           hours = X) %>% 
    filter(hours > 0) # Taking only postive entries as time has to be positive 

print('The data looks like:')
head(excercise_data)
```

## Population Distribution

The distribution of hours spent excercising by 1000000 residents is the population distribution. Let us see how it looks.

```r
set.seed(1234)
options(repr.plot.width=7, repr.plot.height=5)

# Calculate the mean hours spent excercising by the population rounded upto the minutes 

avg_hours = round((mean(excercise_data$hours)),2)

# Plot the Histogram (binwidth = 0.5) of the hours spent by the population excercising per week.

(pop_dist <- excercise_data %>% 
 ggplot(aes(x=hours))+
 geom_histogram(binwidth=0.5, fill = "skyblue", color = 'black', alpha=0.6)+
 labs(x='Hours spent excercising per week ', y = 'Number of residents ')+
 geom_vline(xintercept = avg_hours,color='red')+
 annotate('text',x=6.2,y=100000,label=paste('Mean:',avg_hours))+
 ggtitle('Population distribution'))

```

## Sample distribution 

From the population we take sample of size 100 (randomly choose 100 residents).  Now we can plot the hours of excercise for these 100 residents. This would be our Sample distribution.

```r
set.seed(1234)

# Obtain a sample of 100 residents from the population 

sample_one_dist <- rep_sample_n(excercise_data, size=100);

# Compute the mean hours of the sample, rounded to minutes

sample_one_dist_mean <- round(mean(sample_one_dist$hours),2)

# Plot the histogram plot (binwidth = 0.5) of the sample distribution along with the mean of the sample  

(sample_dist <- sample_one_dist %>% 

 ggplot(aes(hours))+

 geom_histogram(binwidth=0.5, fill = "skyblue", color = 'black', alpha=0.6)+

 labs(x='Hours spent excercising per week ', y = 'Number of residents ')+

 geom_vline(xintercept = sample_one_dist_mean,color='red')+

 annotate('text',x=7,y=15,label=paste('Mean:',sample_one_dist_mean))+

 ggtitle('Sample distribution'))

```

This is a plot from just one sample chosen at random from the population. The mean can be different for different samples. For the sample displayed above the mean is 7.9 Hours. We cannot conclude anything about the population mean just looking at one sample distribution.

# Sampling  distribution 

We obtain 10000 samples each comprising 100 residents. We plot the mean of each of these 10000 samples. This distribution of the means in our case would be called the sampling distribution.


```r
set.seed(1234)

# Create the sampling distribution of 10000 samples of size 100 each

sample_10000_dist <- rep_sample_n(excercise_data, size=100, reps = 10000);

# Compute the mean of each of the sample of size 100 out of the 10000 samples

sample_10000_dist <- sample_10000_dist %>% 
                        group_by(replicate) %>% 
                            summarise(mean_10000_samples=mean(hours))

# Compute the mean and sd of the sampling distribution i.e mean of the 10000 sample means (rounded to minutes) 

mean_sampling_dist <- round(mean(sample_10000_dist$mean_10000_samples),2)
std_sampling_dist <- sd(sample_10000_dist$mean_10000_samples)

# Plot the sampling distribution overlayed by the mean of the sampling distribution 

(sampling_dist <- sample_10000_dist%>% 

 ggplot(aes(mean_10000_samples))+

 geom_histogram(binwidth=0.1, fill = "skyblue", color = 'black', alpha=0.6)+

 labs(x=' Mean Hours of 10000 samples', y = 'Sample Count ')+

 geom_vline(xintercept = mean_sampling_dist, color='Red')+

 annotate('text',x=8.5,y=1000,label=paste('Mean:',mean_sampling_dist))+

 ggtitle('Sampling distribution'))

```

Above plot is plot of the means of the 10000 samples. 
We obtain the mean of the means and we notice it is quite close to the population mean. This is thanks to the Central limit theorem. The central limit theorem states that when an infinite number of successive random samples are taken from a population, the sampling distribution of the means of those samples will become approximately normally distributed with mean μ (same as the population mean) and standard deviation same as the population standard deviation as the sample size (N) becomes larger, irrespective of the shape of the population distribution.

In our case the population distribution is bimodal but the sampling distribution still comes out to be Normal.



# Comparison of three distributions 


```r
# Plot 3 distributions side by side 

options(repr.plot.width=10, repr.plot.height=3)
grid.arrange(pop_dist,sample_dist,sampling_dist, ncol=3)
```

## Summary
